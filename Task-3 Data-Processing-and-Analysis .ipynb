## 1.Employee Performance Bonus Eligibility

```python
# Given employees data
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest = 0
for score in employees.values():
    if score > highest:
        highest = score

# Store top performers in a list
top_names = []

for name in employees:
    if employees[name] == highest:
        top_names.append(name)

# Print the result
print("Top Performers Eligible for Bonus:", top_names, "(Score:", highest, ")")
```

## 2. Search Query Keyword Analysis

```python
# Given query
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.replace(".", "").replace(",", "")

# Split sentence into words
words = query.split()

# Count frequency
count = {}

for word in words:
    if word in count:
        count[word] += 1
    else:
        count[word] = 1

# Print only keywords searched more than once.

result = {}

for word in count:
    if count[word] > 1:
        result[word] = count[word]

print(result)
```

## 3.Sensor Data Validation

```python
# Given Sensor readings 
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid readings
valid_readings = []

# Loop through each reading with its hour index
for hour in range(len(sensor_readings)):
    reading = sensor_readings[hour]
     # for even readings
    if reading % 2 == 0:
        valid_readings.append((hour, reading))  

# Print the valid readings
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## 4. Email Domain Usage Analysis

```python
# Given email list
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Count how many users per domain
domain_count = {}

for email in emails:
    # Get the domain part after '@'
    parts = email.split("@")
    domain = parts[1]
    
    if domain in domain_count:
        domain_count[domain] = domain_count[domain] + 1
    else:
        domain_count[domain] = 1

# Calculate total number of emails
total_users = len(emails)

# Print percentage of each domain
for domain in domain_count:
    percentage = (domain_count[domain] * 100) / total_users
    print(domain + ": " + str(int(percentage)) + "%")
```

## 5.Sales Spike Detection

```python
# Given sales list
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate total and average
total = 0
for sale in sales:
    total = total + sale

average = total / len(sales)
print("Average sales:", average)

# Set spike 
threshold = average * 1.29  
print("Spike threshold:", threshold)

# Check each day for spikes
day = 1
for sale in sales:
    if sale >= threshold:
        print("Day", day, ":", sale)
    day = day + 1
```

## 6.Duplicate User ID Detection

```python
# Given user id list
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

counts = {}  # dictionary to store count

for user in user_ids:
    if user in counts:
        counts[user] += 1  # increase count if already exists
    else:
        counts[user] = 1   # first time seeing the user

# print only duplicates
for user in counts:
    if counts[user] > 1:
        print(user, ":", counts[user], "times")
```
