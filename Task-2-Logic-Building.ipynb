## Problem 1: Unique Words in a Sentence 

```python
sentence = input("Enter sentence: ")

# split into words
words = sentence.split()

# remove duplicates
unique_words = set(words)

# count unique words
count = len(unique_words)

print("Unique words count:", count)
print("Unique words:", unique_words)
```

## Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = ""
highest_salary = 0

# find highest salary
for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

print("Highest Salary:", highest_employee, "-", highest_salary)
```

## Problem 3: Find Maximum and Minimum Values 

```python
numbers = [45, 22, 89, 10, 66]
max_value = numbers[0]
min_value = numbers[0]


for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

## Problem 4: Count Products Above a Price Threshold 

```python
prices = [450, 1200, 899, 1500, 300]
count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

## Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

total_days = len(attendance)

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

## Problem 7: Count Character Frequency 

```python
text = input("Enter a string: ")
char_count = {}

for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print("Character Frequency:", char_count)
```

## Problem 9: Check if a Key Exists in a Dictionary

```python
# If exists

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

name = input("Enter employee name: ")

if name in employees:
    print("Employee exists")
else:
    print("Employee not found")
```

```python
# If not exists

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

name = input("Enter employee name: ")

if name in employees:
    print("Employee exists")
else:
    print("Employee not found")
```

## Problem 10: Calculate Average Marks

```python
marks = [70, 80, 65, 75]

total = sum(marks)
count = len(marks)

average = total / count

print("Average Marks:", average)
```
