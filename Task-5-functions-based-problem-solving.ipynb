# Problem 1: Social Media – Post Engagement Analyzer

### Problem Statement: Create a function to analyze engagement on a social media post.

## Rules
#### Calculate total likes If Total Likes ≥ 1000 → Viral Post
#### Otherwise → Normal Engagement
#### Real-Time Use Social media analytics platforms Marketing dashboards
#### Hint Use function Use loop for calculations
#### Expected Output Format
#### Total Likes: 1050
#### Post Status: Viral Post

```python
def analyze_post_engagement(likes):
    total_likes = 0
    if isinstance(likes, list):
        for like_count in likes:
            total_likes += like_count
    elif isinstance(likes, int):
        total_likes = likes
    else:
        return {"Error": "Invalid input type for likes. Expected int or list of ints."}

    post_status = "Normal Engagement"
    if total_likes >= 1000:
        post_status = "Viral Post"

    return {"Total Likes": total_likes, "Post Status": post_status}
    
result1 = analyze_post_engagement(1050)
print(f"Total Likes: {result1['Total Likes']}")
print(f"Post Status: {result1['Post Status']}")
print("\n")

result2 = analyze_post_engagement([500, 300, 250, 50])
print(f"Total Likes: {result2['Total Likes']}")
print(f"Post Status: {result2['Post Status']}")
print("\n")

result3 = analyze_post_engagement(700)
print(f"Total Likes: {result3['Total Likes']}")
print(f"Post Status: {result3['Post Status']}")
```

# Problem 3: Agriculture – Rainfall Adequacy Checker

### Problem Statement: Create a function that analyzes seasonal rainfall data.

## Rules Calculate average rainfall
#### Average ≥ required level → Adequate Rainfall
#### Otherwise → Inadequate Rainfall
#### Real-Time Use
#### Climate monitoring systems Irrigation planning tools
#### Hint Use function
#### Use loop and math operations
#### Expected Output Format
#### Average Rainfall: 72
#### Rainfall Status: Adequate Rainfall

```python
def analyze_rainfall_adequacy(rainfall_data, required_level):
    if not rainfall_data:
        return {"Error": "Rainfall data cannot be empty."}

    total_rainfall = 0
    for rain_amount in rainfall_data:
        total_rainfall += rain_amount

    average_rainfall = total_rainfall / len(rainfall_data)

    rainfall_status = "Inadequate Rainfall"
    if average_rainfall >= required_level:
        rainfall_status = "Adequate Rainfall"
    return {"Average Rainfall": round(average_rainfall), "Rainfall Status": rainfall_status}
    
rainfall_season1 = [60, 70, 80, 90]
required_level1 = 70
result1 = analyze_rainfall_adequacy(rainfall_season1, required_level1)
print(f"Average Rainfall: {result1['Average Rainfall']}")
print(f"Rainfall Status: {result1['Rainfall Status']}")
print("\n")

rainfall_season2 = [40, 50, 60, 55]
required_level2 = 70
result2 = analyze_rainfall_adequacy(rainfall_season2, required_level2)
print(f"Average Rainfall: {result2['Average Rainfall']}")
print(f"Rainfall Status: {result2['Rainfall Status']}")
```

# Problem 4: Social Media – Duplicate Account Detection

### Problem Statement: Create a function to detect duplicate usernames in a system.

## Rules: If duplicates exist → Flag accounts
#### Else → Accounts are unique
#### Real-Time Use
#### Platform moderation Fraud detection systems
#### Hint
#### Use function
#### Use set data structure
#### Expected Output Format
#### Duplicate Accounts Found: Yes

```python
def detect_duplicate_accounts(usernames):
    seen_usernames = set()
    duplicates = set()

    for username in usernames:
        if username in seen_usernames:
            duplicates.add(username)
        else:
            seen_usernames.add(username)

    if duplicates:
        return {"Duplicate Accounts Found": "Yes", "Duplicates": list(duplicates)}
    else:
        return {"Duplicate Accounts Found": "No"}
        
users1 = ["Hemant", "Prajakta", "Paras", "Prajakta", "Hemant"]
result1 = detect_duplicate_accounts(users1)
print(f"Users: {users1}")
print(f"Duplicate Accounts Found: {result1['Duplicate Accounts Found']}")

if 'Duplicates' in result1:
    print(f"Duplicates: {result1['Duplicates']}")
print("\n")

users2 = ["Pranav", "Suyog", "Ranbir"]
result2 = detect_duplicate_accounts(users2)
print(f"Users: {users2}")
print(f"Duplicate Accounts Found: {result2['Duplicate Accounts Found']}")
print("\n")

users3 = ["Kranti", "Dipali", "Aditya", "Kranti", "Dipali", "Palak", "Aditya"]
result3 = detect_duplicate_accounts(users3)
print(f"Users: {users3}")
print(f"Duplicate Accounts Found: {result3['Duplicate Accounts Found']}")
if 'Duplicates' in result3:
    print(f"Duplicates: {result3['Duplicates']}")
```

# Problem 5: Healthcare – Appointment Eligibility Checker

### Problem Statement: Create a function to check whether a patient is eligible for an appointment slot.

## Rules
#### Age ≥ 18 → Eligible
#### Age < 18 → Not Eligible
#### Real-Time Use
#### Hospital appointment systems
#### Healthcare portals

```python
def check_appointment_eligibility(age):
    """
    Checks a patient's eligibility for an appointment based on age.

    Args:
        age (int): The age of the patient.

    Returns:
        dict: A dictionary containing 'Patient Age' and 'Eligibility Status'.
    """
    eligibility_status = "Not Eligible"
    if age >= 18:
        eligibility_status = "Eligible"

    return {"Patient Age": age, "Eligibility Status": eligibility_status}

# Example Usage:
# Test with an eligible patient
patient_age1 = 21
result1 = check_appointment_eligibility(patient_age1)
print(f"Patient Age: {result1['Patient Age']}")
print(f"Eligibility Status: {result1['Eligibility Status']}")
print("\n")
# Test with a not eligible patient
patient_age2 = 16
result2 = check_appointment_eligibility(patient_age2)
print(f"Patient Age: {result2['Patient Age']}")
print(f"Eligibility Status: {result2['Eligibility Status']}")
```

# Problem 6: Agriculture – Premium Crop Price Filter

### Problem Statement: Create a function that filters crop prices above a premium threshold.

## Rules
#### Price > ₹2000 → Premium Crop
#### Else → Regular Crop
#### Real-Time Use
#### Market analytics
#### Export pricing systems

```python
def filter_premium_crops(crop_prices):
    premium_crops = []
    # Use list filtering logic as per hint
    for price in crop_prices:
        if price > 2000:
            premium_crops.append(price)

    return {"Premium Crops": premium_crops}
prices1 = [1500, 2500, 1800, 3200, 1950]
result1 = filter_premium_crops(prices1)
print(f"Original Prices: {prices1}")
print(f"Premium Crops: {result1['Premium Crops']}")
print("\n")

prices2 = [1000, 1200, 800, 1999]
result2 = filter_premium_crops(prices2)
print(f"Original Prices: {prices2}")
print(f"Premium Crops: {result2['Premium Crops']}")
print("\n")

prices3 = [2100, 3000, 4500]
result3 = filter_premium_crops(prices3)
print(f"Original Prices: {prices3}")
print(f"Premium Crops: {result3['Premium Crops']}")
```

# Problem 7: System Monitoring – Application Health Checker

### Problem Statement: Create a function to evaluate application health based on error count.

## Rules
#### Errors = 0 → Healthy
#### Errors ≤ 5 → Minor Issues
#### Errors > 5 → Critical Issues
#### Real-Time Use
#### System monitoring tools
#### IT operations dashboards

```python
def check_application_health(error_count):
    system_status = "Healthy"
    if error_count > 5:
        system_status = "Critical Issues"
    elif error_count > 0 and error_count <= 5: # This covers 1 to 5 errors
        system_status = "Minor Issues"

    return {"Error Count": error_count, "System Status": system_status}

error_count1 = 7
result1 = check_application_health(error_count1)
print(f"Error Count: {result1['Error Count']}")
print(f"System Status: {result1['System Status']}")
print("\n")

error_count2 = 3
result2 = check_application_health(error_count2)
print(f"Error Count: {result2['Error Count']}")
print(f"System Status: {result2['System Status']}")
print("\n")

error_count3 = 0
result3 = check_application_health(error_count3)
print(f"Error Count: {result3['Error Count']}")
print(f"System Status: {result3['System Status']}")
print("\n")

error_count4 = 5
result4 = check_application_health(error_count4)
print(f"Error Count: {result4['Error Count']}")
print(f"System Status: {result4['System Status']}")
```

# Problem 8: Banking – Daily Transaction Limit Checker

### Problem Statement: Create a function to validate daily transaction amount.

## Rules
#### Daily limit = ₹50,000
#### Amount within limit → Approved
#### Amount exceeds limit → Rejected
#### Real-Time Use
#### Banking systems
#### Fraud prevention systems

```python
def check_transaction_limit(transaction_amount):
    DAILY_LIMIT = 50000
    transaction_status = "Rejected"

    if transaction_amount <= DAILY_LIMIT:
        transaction_status = "Approved"

    return {"Transaction Amount": transaction_amount, "Transaction Status": transaction_status}

transaction1 = 60000
result1 = check_transaction_limit(transaction1)
print(f"Transaction Amount: {result1['Transaction Amount']}")
print(f"Transaction Status: {result1['Transaction Status']}")
print("\n")

transaction2 = 45000
result2 = check_transaction_limit(transaction2)
print(f"Transaction Amount: {result2['Transaction Amount']}")
print(f"Transaction Status: {result2['Transaction Status']}")
print("\n")

transaction3 = 50000
result3 = check_transaction_limit(transaction3)
print(f"Transaction Amount: {result3['Transaction Amount']}")
print(f"Transaction Status: {result3['Transaction Status']}")
```

# Problem 9: E-Learning – Student Attendance Eligibility System

### Problem Statement: Create a function to calculate attendance percentage and check exam eligibility.

## Rules
#### Attendance ≥ 75% → Eligible
#### Attendance < 75% → Not Eligible
#### Real-Time Use
#### Learning management systems
#### Internship tracking platforms

```python
def check_attendance_eligibility(attendance_records):
    if not attendance_records:
        return {"Error": "Attendance records cannot be empty."}

    total_classes = len(attendance_records)
    classes_attended = sum(attendance_records)

    attendance_percentage = (classes_attended / total_classes) * 100

    exam_eligibility = "Not Eligible"
    if attendance_percentage >= 75:
        exam_eligibility = "Eligible"

    return {"Attendance Percentage": round(attendance_percentage, 1), "Exam Eligibility": exam_eligibility}

student1_attendance = [True, True, True, True, False, True, True, True, True, False]
result1 = check_attendance_eligibility(student1_attendance)
print(f"Attendance Percentage: {result1['Attendance Percentage']}")
print(f"Exam Eligibility: {result1['Exam Eligibility']}")
print("\n")

student2_attendance = [True, True, True, False, False, True, False, False, True, False]
result2 = check_attendance_eligibility(student2_attendance)
print(f"Attendance Percentage: {result2['Attendance Percentage']}")
print(f"Exam Eligibility: {result2['Exam Eligibility']}")
print("\n")

student3_attendance = [True, True, True, True, True, True, True, False, False, False] # 7 out of 10
result3 = check_attendance_eligibility(student3_attendance)
print(f"Attendance Percentage: {result3['Attendance Percentage']}")
print(f"Exam Eligibility: {result3['Exam Eligibility']}")
print("\n")

student4_attendance = [True, True, True, True, True, True, False, False, False, False] # 6 out of 10
result4 = check_attendance_eligibility(student4_attendance)
print(f"Attendance Percentage: {result4['Attendance Percentage']}")
print(f"Exam Eligibility: {result4['Exam Eligibility']}")
```

# Problem 10: Smart Electricity Bill Analyzer

### Problem Statement: You are building a smart electricity billing system for residential users. Create a function that calculates the electricity bill and classifies usage level based on units consumed.

## Billing Rules
#### First 100 units → ₹3 per unit
#### Next 100 units (101–200) → ₹5 per unit
#### Above 200 units → ₹7 per unit
#### Usage Classification
#### Total bill < ₹500 → Low Usage
#### ₹500 – ₹1500 → Moderate Usage
#### ₹1500 → High Usage

```python
def calculate_electricity_bill(units_consumed):

    total_bill = 0
    if units_consumed <= 100:
        total_bill = units_consumed * 3
    elif units_consumed <= 200:
        total_bill = (100 * 3) + ((units_consumed - 100) * 5)
    else:
        total_bill = (100 * 3) + (100 * 5) + ((units_consumed - 200) * 7)
    usage_classification = ""
    if total_bill < 500:
        usage_classification = "Low Usage"
    elif 500 <= total_bill <= 1500:
        usage_classification = "Moderate Usage"
    else:
        usage_classification = "High Usage"

    return {"Units Consumed": units_consumed, "Total Bill": total_bill, "Usage Classification": usage_classification}

units1 = 50
result1 = calculate_electricity_bill(units1)
print(f"Units Consumed: {result1['Units Consumed']}")
print(f"Total Bill: ₹{result1['Total Bill']}")
print(f"Usage Classification: {result1['Usage Classification']}")
print("\n")

units2 = 150
result2 = calculate_electricity_bill(units2)
print(f"Units Consumed: {result2['Units Consumed']}")
print(f"Total Bill: ₹{result2['Total Bill']}")
print(f"Usage Classification: {result2['Usage Classification']}")
print("\n")

units3 = 301
result3 = calculate_electricity_bill(units3)
print(f"Units Consumed: {result3['Units Consumed']}")
print(f"Total Bill: ₹{result3['Total Bill']}")
print(f"Usage Classification: {result3['Usage Classification']}")
print("\n")

units4 = 200
result4 = calculate_electricity_bill(units4)
print(f"Units Consumed: {result4['Units Consumed']}")
print(f"Total Bill: ₹{result4['Total Bill']}")
print(f"Usage Classification: {result4['Usage Classification']}")
print("\n")

units5 = 100
result5 = calculate_electricity_bill(units5)
print(f"Units Consumed: {result5['Units Consumed']}")
print(f"Total Bill: ₹{result5['Total Bill']}")
print(f"Usage Classification: {result5['Usage Classification']}")
```
