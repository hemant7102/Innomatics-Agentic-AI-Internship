# **Task 1: Conceptual Understanding**

**Q1. Difference between "Love" and "love" in NLP**

**Answer:** In Natural Language Processing, "Love" and "love" are treated as two different tokens when the text is in its raw form. This happens because most NLP systems are case-sensitive by default. As a result, the same word written in different cases is considered separate, which unnecessarily increases the vocabulary size and can affect model efficiency. To avoid this issue, a common preprocessing step is to convert all text into lowercase. After this normalization, both "Love" and "love" are treated as the same word, making the data more consistent and easier for models to process.


**Q2. What happens if stopwords are not removed?**

**Answer:** If stopwords are not removed during preprocessing, the dataset tends to become noisy. Stopwords are very common words such as "is", "the", and "and" that usually do not carry significant meaning. Keeping them increases the size of the data without adding much value, which can reduce model performance and slow down processing.


**Q3. When removing stopwords can be harmful**

**Answer:** There are certain real-world scenarios where removing stopwords can lead to loss of important information. One such case is sentiment analysis. For example, in the sentence "I am not happy", the word "not" plays a crucial role in conveying a negative sentiment. If stopwords are removed, the sentence becomes "happy", which completely changes the meaning and leads to incorrect interpretation.


**Q4. Difference between stemming and lemmatization**

**Answer:** Stemming and lemmatization are both techniques used to reduce words to their base or root form, but they work differently. Stemming is a simple, rule-based approach that removes prefixes or suffixes from words. Because of this, it may sometimes produce outputs that are not actual words. For example, "running" might be reduced to "runn".

On the other hand, lemmatization is a more advanced method that uses a dictionary and linguistic rules to convert words into their meaningful base forms. It ensures that the output is always a valid word. For instance, "running" is correctly reduced to "run". While stemming is faster and simpler, lemmatization is more accurate and provides better results in most NLP tasks.

# **Task 2: Advanced Preprocessing Function**

```python
import re
import nltk
from nltk.tokenize import word_tokenize
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ['no', 'not']
    ]

    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```


# **Task 3: Stress Testing**

```python
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://google.com now!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for text in sample_inputs:
    tokens, sentence = preprocess_text(text)
    results.append((text, tokens, sentence))

for r in results:
    print("\nOriginal:", r[0])
    print("Tokens:", r[1])
    print("Cleaned:", r[2])
```

# **Task 4: Token Analytics**

```python
def token_analytics(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0

    return total_tokens, unique_tokens, avg_length

for text, tokens, sentence in results:
    total, unique, avg = token_analytics(tokens)

    print("\nSentence:", text)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Avg Token Length:", round(avg, 2))
```

# **Task 5: Frequency Analysis**

```python
from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("\nTop 10 frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

# **Task 6: Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
output = full_pipeline(sample_inputs)
print(output)
```

# **Task 7: Error Handling**

```python
def preprocess_text(text):

    if not text or text.strip() == "":
        return [], ""

    import re

    text = text.lower()

    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)

    text = re.sub(r'[^a-z\s]', '', text)

    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    if not tokens:
        return [], ""

    return tokens, " ".join(tokens)

test_cases = ["", "😍😍😂😂", "1234567890"]

for text in test_cases:
    tokens, sentence = preprocess_text(text)

    print("\nOriginal Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", sentence)
```
