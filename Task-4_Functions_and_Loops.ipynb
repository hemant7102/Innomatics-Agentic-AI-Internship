# Problem 1: Mobile Recharge Validation System

## Problem Statement
#### Create a function to validate a mobile recharge request.

## Rules:
#### Recharge amount must be ≥ ₹50
#### Valid plans: 199, 299, 399, 599
#### User should be allowed to retry if invalid

```python
# Function to validate the recharge amount

def validate_recharge(amount):
    # List of valid plans
    valid_plans=[199,299,399,599]
    # Minimum recharge amount must be ₹50
    if amount<50:
        print("Minimum Recharge amount is 50")
    # Amount must match one of the valid plans
    elif amount not in valid_plans:
        print("Invalid Plan")
    else:
        return True
# loop to allow user retry until valid input
while True:
    recharge_amount=int(input("Enter Recharge Amount(199/299/399/599):"))
    if validate_recharge(recharge_amount):
        print("Recharge Successful!")
        #  Exit loop if recharge is valid
        break
    else:
        print("Try Again")   
```

# Problem 2: Inventory Reorder Alert System

## Problem Statement

### Given a dictionary of products and stock quantities:
#### Print Reorder Alert if stock < 15
#### Print Stock OK otherwise

```python
# check inventory and print reorder alerts

def inventory_reorder_alert(inventory):
    for product,stock in inventory.items():
        #  Check if stock is below 15
        if stock<15:
             print(f"{product} -> Reorder Alert (Stock: {stock})")
        else:
            print(f"{product} -> Stock OK (Stock: {stock})")

# Inventory dictionary
inventory = {
    "Laptop": 15,
    "Smartphone": 40,
    "Headphones": 5,
    "Keyboard": 50,
    "Mouse": 6,
    "Monitor": 20,
    "USB Cable": 10,
    "Power Bank": 30
}

# Function call
inventory_reorder_alert(inventory)
```

# Problem 3: Student Result Processing System

## Problem Statement

### Create a function that:
#### Accepts a list of student marks
#### Calculates average marks
#### Prints Pass if average ≥ 50 else Fail


```python
# Function to process student result
def student_result(student_marks):    
    # Handle empty list case
    if len(student_marks) == 0:
        return "No marks provided"    
    total = 0
    
    # Calculate total using loop
    for marks in student_marks:
        total += marks
    
    # Calculate average
    average = total / len(student_marks)
    print(f"Average: {average}")    
    # Return result based on average
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example usage
student_marks = [60, 45, 70, 55, 40]

result = student_result(student_marks)
print("Result:", result)
```

# Problem 4: Cab Fare Estimator with Retry Option

## Problem Statement

### Create a cab fare calculator:
#### Base fare = ₹50
#### ₹12 per km
#### Peak hour → 25% extra
#### Allow retry using while loop

```python
# function for fare calculation
def calculate_fare(distance,is_peak):
    # cab fare's
    base_fare = 50
    per_km_rate = 12
    # Calculate basic fare
    total_fare = base_fare + (distance * per_km_rate)
    
    if is_peak:
        total_fare += total_fare * 0.25
    return total_fare

#loop to allow user to retry fare calculation
while True:
    try:
        distance = float(input("Enter distance in km: "))
        
        if distance <= 0:
            print("Distance must be positive.\n")
            continue
        # Ask user if it is peak hour
        peak_input = input("Is it peak hour? (yes/no): ").lower()

        
        if peak_input not in ["yes", "no"]:
            print("Invalid input for peak hour.\n")
            continue
            
         # Exit loop if user does not want to retry
        is_peak = True if peak_input == "yes" else False
        
        fare = calculate_fare(distance, is_peak)
        
        print(f"Total Fare: ₹{fare:.2f}")
        
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using the fare estimator.")
            break
    
    except ValueError:
        print("Invalid input. Please enter numeric value for distance.\n")
```

# Problem 5: Employee Attendance Eligibility Checker

## Problem Statement

### Create a function that:
#### Accepts attendance list ("P", "A")
#### Calculates attendance percentage
#### Returns Eligible if ≥ 75% else Not Eligible

```python
# Function to check attendance eligibility
def check_attendance(attendance_list):
    
    total_days = len(attendance_list)   # Total number of days
    present_days = 0                    # Counter for present days
    
    # Loop through each day's attendance
    for day in attendance_list:
        
        # If student was present, increase counter
        if day == "P":
            present_days += 1
    
    # Calculate attendance percentage
    attendance_percentage = (present_days / total_days) * 100
    
    # Check eligibility condition
    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example attendance record
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

# Call function and store result
result = check_attendance(attendance)

# Print result
print("Attendance Result:", result)
```

# Problem 6: Password Strength Checker

## Problem Statement
### Create a function to check password strength.

## Rules:
#### Minimum length = 8
#### Must contain at least one digit
#### Must contain at least one special character (@#$)

```python
# Function to check password strength
def check_password_strength(password):
    
    # Rule 1: Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum length should be 8)"
    
    has_digit = False          # To check if password contains a number
    has_special = False        # To check if password contains special character
    
    # Loop through each character in password
    for char in password:
        
        # Check if character is a digit
        if char.isdigit():
            has_digit = True
        
        # Check if character is one of required special symbols
        if char in "@#$":
            has_special = True
    
    # Check all conditions
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"

# Example usage
password = input("Enter your password: ")

result = check_password_strength(password)

print(result)
```
